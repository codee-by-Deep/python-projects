```python
movies.groupby('Genre')['IMDB_Rating'].max().sort_values(ascending = False).head(10)
```

```python
movies.head()
```

```python
movies['Released_Year'].value_counts().head()
```

```python
Director_counts = movies['Director'].value_counts()
```

```python
top5_directors = Director_counts[Director_counts >= 5].index
```

```python
movies[movies['Director'].isin(top5_directors)].groupby('Director')['IMDB_Rating'].mean().sort_values(ascending = False).head()
```

```python
movies['Runtime'] = movies["Runtime"].str.replace("min",'')
```

```python
movies["Runtime"] = movies['Runtime'].astype(float)
```

```python
movies['Runtime']
```

```python
pd.concat([movies['Star1'],movies['Star2'],movies['Star3'],movies['Star4']]).value_counts().head()
```

```python
movies.info()
```

```python
movies["Released_Year"] = pd.to_numeric(movies["Released_Year"], errors="coerce")
```

```python
movies["Released_Year"] = movies["Released_Year"].astype("Int64")
```

```python
movies["Decade"] = (movies["Released_Year"] // 10 * 10)
```

```python
movies['Gross'] = movies['Gross'].astype(float)
```

```python
Grossperdecade = movies.groupby('Decade')['Gross'].mean().sort_values(ascending = False).head().astype(int)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(["Drama","Action","Comedy"],[85,30,25])
plt.title("Movie by Genre")
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.show()
```

```python
genre_counts = movies["Genre"].value_counts().head(10)
```

```python
genre_counts
```

```python
plt.bar(genre_counts.index, genre_counts.values)
plt.title("Top 10 Genre in IMDB Top 1000")
plt.xlabel("Genre")
plt.ylabel("Number of Movies")
plt.xticks(rotation = 45)
plt.show()
```

```python
Director_counts = movies["Director"].value_counts().head(10)
```

```python
Director_counts
```

```python
plt.bar(Director_counts.index,Director_counts.values)
plt.title("Top 10 Directors in IMDB Top 1000")
plt.xlabel("Director")
plt.ylabel("Number of Movies")
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.plot(Grossperdecade.index, Grossperdecade.values, marker = "o", color = "green")
plt.title("Average Gross by Decade")
plt.xlabel("Decade")
plt.ylabel("Average Gross ($)")
plt.ticklabel_format(style="plain", axis="y")
plt.show()
```

```python
top10 = movies[["Series_Title",'IMDB_Rating']].sort_values('IMDB_Rating',ascending = False).head(10)

plt.barh(top10['Series_Title'],top10['IMDB_Rating'], color = 'orange')
plt.title("Top 10 Movies by IMDB Rating")
plt.xlabel("IMDB Rating")
plt.tight_layout()
plt.show()
```
